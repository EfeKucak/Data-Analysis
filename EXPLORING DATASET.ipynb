```python
import numpy as np

# Yaş aralıkları ve katılımcı sayıları
age_groups = {
    "Under 18 years old": 2568,
    "18-24 years old": 14098,
    "25-34 years old": 23911,
    "35-44 years old": 14942,
    "45-54 years old": 6249,
    "55-64 years old": 2575,
    "65 years or older": 772
}

# Her yaş grubuna temsilci yaş atama (ortalama yaşı kullanıyoruz)
age_midpoints = {
    "Under 18 years old": 16,   # 18 yaş altı için tahmini bir yaş
    "18-24 years old": 21,      # (18+24)/2 = 21
    "25-34 years old": 29.5,    # (25+34)/2 = 29.5
    "35-44 years old": 39.5,    # (35+44)/2 = 39.5
    "45-54 years old": 49.5,    # (45+54)/2 = 49.5
    "55-64 years old": 59.5,    # (55+64)/2 = 59.5
    "65 years or older": 70     # 65+ için tahmini yaş
}

# Ağırlıklı ortalama hesaplama
total_weighted_age = sum(age_midpoints[age] * count for age, count in age_groups.items())
total_count = sum(age_groups.values())

mean_age = total_weighted_age / total_count
print(f"Mean Age: {mean_age:.2f}")
```
