# Task 2: Analyze Characteristics of Duplicates

Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.

Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.

```python
sub_cols=["MainBranch", "Employment","RemoteWork"]

duplicates=df[df.duplicated(subset=sub_cols , keep=False)]               # RETURNS ALL THE DUPLICATE ROWS BASED ON GIVEN COLUMNS
```

```python
duplicates
```

```python
# Count identical values across duplicate rows

identical_counts = (duplicates.groupby(sub_cols).nunique() == 1).sum()

# Display columns where values are frequently identical
print(identical_counts.sort_values(ascending=False))
```

# Task 3: Visualize Duplicates Distribution
Create visualizations to show the distribution of duplicates across different categories.
Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.

```python
country_count=duplicate_rows["Country"].value_counts()
emp_count=duplicate_rows["Employment"].value_counts()
```

```python
plt.figure(figsize=(12,6))

# Country distribution
plt.bar(country_count.index, country_count.values, label="Country", color="green", width=0.4, align='center')
plt.xticks(rotation=90)
plt.show()

# Employment pie chart
plt.pie(emp_count, labels=emp_count.index,autopct='%1.1f%%') 

plt.show()


```
